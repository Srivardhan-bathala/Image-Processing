```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread("car.jpg")
#Converting from BGR to RGB color
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#Convert to HSV color mapping
#note: HSV and HLS color mappings are a close representation of how human beings perceive color
hsv_img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img)
```

```python
#Concert to HLS color mapping
hls_img = cv2.cvtColor(img,cv2.COLOR_RGB2HLS)
plt.imshow(hls_img)
```
