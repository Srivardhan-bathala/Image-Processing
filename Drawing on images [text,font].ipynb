```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#creating a blank image
img = np.zeros(((512,512,3)))
```

```python
plt.imshow(img)
```

```python
#code for adding font on blank image
font = cv2.FONT_ITALIC
cv2.putText(img,text = "good morning",org = (30,400),fontFace = font, fontScale =2 , color = (255,255,255),thickness = 3,
            lineType = cv2.LINE_AA)
plt.imshow(img)
           
```

```python
p_img = np.zeros(((512,512,3)))
```

```python
#creating an array of points which are the vertices of the polygon
vertices = np.array([[100,400],[200,200],[300,100],[450,450]])
```

```python
#our blank image is in three dimensions, hence this would show up as an error on cv2 because our array of points are in two dimensions
#this code adds another dimension
pt = vertices.reshape(-1,1,2)
```

```python
#code for drawing polygons on blank image
poly_img = cv2.polylines(p_img,[pt],isClosed = True,color = (0,255,255),thickness = 10)
plt.imshow(poly_img)
```
