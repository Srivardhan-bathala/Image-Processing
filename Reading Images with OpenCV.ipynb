```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#reading image
img = cv2.imread("car.jpg")
```

```python
#displaying image
plt.imshow(img)
#its blueish because openCV uses bgr not rgb
```

```python
#converting color channels
new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
```

```python
#converting to a gray scale
gray_img = cv2.imread("car.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(gray_img,cmap='gray')
```
