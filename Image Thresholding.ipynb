```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Reading the image in grayscale
img = cv2.imread('parrot.jpg',0)
#Since matplotlib and open cv read in differnt color schemes, we need to use cmap parameter to display in grayscale
plt.imshow(img,cmap='gray')
```

```python
img.max()
```

```python
thresh,thresh_img = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```
