```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Reading images
img1 = cv2.imread('dog2.jpeg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('car.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
img3 = cv2.imread('ms.jpg')
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
#Blending images of the same size
blend = cv2.addWeighted(src1 =img1,alpha = 0.5,src2 = img2,beta =0.5,gamma= 0)
plt.imshow(blend)
```

```python
plt.imshow(img3)
```

```python
#Overlaying small image on top of large image
x_start = 0
y_start = 0
x_end = x_start + img2.shape[1]
y_end = y_start + img2.shape[0]
img3[y_start:y_end,x_start:x_end] = img2
```

```python
plt.imshow(img3)
```
