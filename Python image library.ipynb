```python
#importing PIL package
from PIL import Image
```

```python
#reading the image
img = Image.open('bg.jpg')
```

```python
#converting image to a numpy array
img_arr = np.array(img)
```

```python
#displaying image using matplotlib
plt.imshow(img_arr)
```

```python
#removing red layer
img_red = img_arr[:,:,0]
```

```python
plt.imshow(img_red)
```

```python
#showing image in grayscale
plt.imshow(img_red,cmap='gray')
```

```python
#removing green layer
img_green = img_arr[:,:,1]
plt.imshow(img_green)
```

```python
#showing image in grayscale
plt.imshow(img_green,cmap='gray')
```

```python
#removing blue layer
img_blue = img_arr[:,:,2]
plt.imshow(img_blue)
```

```python
#showing image in grayscale
plt.imshow(img_blue,cmap='gray')
```

```python
#displaying image with red and blue channel
img_noGreen = img_arr.copy()
img_noGreen[:,:,1] = 0
plt.imshow(img_noGreen)
```

```python
#displaying image with green and blue channel
img_noRed = img_arr.copy()
img_noRed[:,:,0] = 0
plt.imshow(img_noRed)
```

```python
#displaying image with red and green channel
img_noBlue = img_arr.copy()
img_noBlue[:,:,2] = 0
plt.imshow(img_noBlue)
```
